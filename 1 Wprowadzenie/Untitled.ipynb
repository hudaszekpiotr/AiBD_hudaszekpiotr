## Zadanie 3
Funkcja $f(x) = x^2+5$

x>-1 oraz x<1

x>-6 oraz x<6

x>0 oraz x<5

Zapoznanie się z Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np


def fun(x):
    return x**2+5

x = np.linspace(-1,1,100)
plt.plot(x,fun(x))
plt.title("y=x^2+5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend("y")
plt.show()

x = np.linspace(-6,6,100)
plt.plot(x,fun(x))
plt.title("y=x^2+5")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

x = np.linspace(0,5,100)
plt.title("y=x^2+5")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,fun(x))
plt.show()
```
